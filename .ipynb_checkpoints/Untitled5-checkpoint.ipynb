```python
from matplotlib.pyplot import imread
import matplotlib.pyplot as plt
from matplotlib.pyplot import imshow
import numpy as np

# 
im = imread('/Users/murakamimanabu/Downloads/girl.jpg')

imshow(im)
plt.title("original RGB image")
plt.show()

# 
r_channel = im[:, :, 0]
g_channel = im[:, :, 1]
b_channel = im[:, :, 2]

fig = plt.figure(figsize=(15, 3))
for i, c in zip(range(3), 'RGB'):
    ax = fig.add_subplot(1, 3, i + 1)
    mappable = ax.imshow(im[:, :, i], vmin=0, vmax=255)
    plt.colorbar(mappable, ax=ax)
    ax.set_title(f'{c} channel')
plt.show()

# RGB
im2 = np.zeros((2, 4, 3), dtype=np.uint8)   #  zero  zeros 
im2[0, 1, 0] = 255  # 1

print("Red channel (0):")
print(im2[:, :, 0])
print("Green channel (1):")
print(im2[:, :, 1])
print("Blue channel (2):")
print(im2[:, :, 2])

imshow(im2)
plt.axis('off')
plt.show()

# RGBA
# 3, 2, 4RGBA
im_rgba = np.zeros((3, 2, 4), dtype=np.uint8)   #   zeros
im_rgba[0, 0, 1] = 255   # G
print(im_rgba)

plt.imshow(im_rgba)
plt.axis('off')
plt.show()

imshow(im)
plt.axis('off')
plt.show()
```

```python
import cv2 as cv
import matplotlib.pyplot as plt
from pathlib import Path

img_path = Path('/Users/murakamimanabu/Downloads/girl.jpg')
img = cv.imread(str(img_path))
if img is None:
    raise FileNotFoundError(f': {img_path}')

img_rgb = cv.cvtColor(img, cv.COLOR_BGR2RGB)
plt.imshow(img_rgb); plt.axis('off'); plt.show()
```
